```python
import pandas as pd
```

```python
#Leitura dos aarquivos
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```python
#Juntando todos os arquivos
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
#Exibindo 5 linhas de amostra do bd
df.sample(5)
```

```python
#Alterando tipo de dado da coluna LojaID
df["LojaID"] = df["LojaID"].astype("object")
```

```python
#Confirmando os tipos de dados
df.dtypes
```

```python
#Consultando linhas com valores faltantes
df.isnull().sum()
```

```python
#Substituindo os valores nulos pela m√©dia
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
```

```python
#Substituindo os valores nulos por zero
df["Vendas"].fillna(0,inplace=True)
```

```python
#Apagando as linhas com os valores nulos
df.dropna(inplace=True)
```

```python
#Apagando as linhas com valores nulos com base em 1 coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
#Removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
```

```python
#Criando coluna nova de receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
```

```python
df.head()
```

```python
#Retornado a maior receita
df["Receita"].max()
```

```python
#Retornado a menor receita
df["Receita"].min()
```

```python
#nlargest top 3 melhores receita
df.nlargest(3, "Receita")
```

```python
#nsamllest top 3 piores receita
df.nsmallest(3, "Receita")
```

```python
#Agrupamento por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#Ordenando o conjunto de dados
df.sort_values("Receita", ascending=False).head(10)
```
